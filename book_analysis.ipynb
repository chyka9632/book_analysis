# Finding the list of sentences where the word "love" occurs

### The non-dynamic approach

```python
pattern = re.compile("[a-zA-Z ,]* love [a-zA-Z ,]*")
re.findall(pattern, book)
```

### The dynamic approach using negation (^),this is like saying give me everything after and before the word "love" but stop at "."

```python
pattern = re.compile("[^.]* love [^.]*")
re.findall(pattern, book)
```

### Adding the fullstop after the last word and removing space from the start.

```python
pattern = re.compile("[A-Z]{1}[^.]* love [^.]*.")
re.findall(pattern, book)
```

### When we apply the "len()" fnx in the above code it gave 49 however that doesn't reflect all the instances where love occur, we need to include instances where love is preceded and succeeded by characters other than space.

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]love[^a-zA-Z][^.]*.")
len(re.findall(pattern, book))
```

# Sentiment Analysis: Finding the most positive and most negative chapter

### An example using random texts

```python
from nltk.sentiment import SentimentIntensityAnalyzer
 
```

```python
analyzer =  SentimentIntensityAnalyzer()
```

```python
scores = analyzer.polarity_scores("Oh! How beautiful is Spring. The longer daylight,the blooms and the freshness of the air. ")
```

```python
scores
```

```python
if scores["pos"]>scores["neg"]:
    print("It is a positive text")
else:
    print("The text is negative")
```

```python
 analyzer.polarity_scores(book) # The above was an example using random text, now we're back to our book.
```

### Chapter sentiment analysis

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)[1:]
```

```python
for nr, chapter in enumerate (chapters):
    sentiment_scores = analyzer.polarity_scores(chapter)
    print(nr + 1, sentiment_scores) # The nr is just used to provide chapter numbering accordinly
    
```

# A function to find the occurence of any word

```python
def find(w):
    pattern = re.compile("[a-zA-z]+")
    words = re.findall(pattern, book.lower())
    d = {}
    for word in words:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
    try:
        return d[w]

    except:
        return f'The book does not contain the word "{w}"'

        
        
```
